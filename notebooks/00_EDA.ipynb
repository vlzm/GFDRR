```python
import requests
import pandas as pd

DISCOVERY_URL = "https://gbfs.citibikenyc.com/gbfs/2.3/gbfs.json"
LANG = "en"

def get_json(url: str, timeout: int = 15) -> dict:
    r = requests.get(url, timeout=timeout)
    r.raise_for_status()
    return r.json()

# 1) discovery: получаем реальные URL фидов
discovery = get_json(DISCOVERY_URL)
feeds = {f["name"]: f["url"] for f in discovery["data"][LANG]["feeds"]}

# 2) базовые фиды: география/метаданные станций + текущий статус
station_info = get_json(feeds["station_information"])["data"]["stations"]
station_status = get_json(feeds["station_status"])["data"]["stations"]

df_info = pd.DataFrame(station_info)
df_status = pd.DataFrame(station_status)

# 3) join по station_id
df = df_info.merge(df_status, on="station_id", how="left", suffixes=("_info", "_status"))

# пример: топ-15 станций по доступным велосипедам прямо сейчас
cols = ["station_id", "name", "lat", "lon", "num_bikes_available", "num_docks_available", "is_installed", "is_renting", "is_returning", "last_reported"]
print(df[cols].sort_values("num_bikes_available", ascending=False).head(15))
```
